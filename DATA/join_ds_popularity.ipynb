```python
import pandas as pd
import numpy as np

names_data = ["dataset_05-2024", "dataset_06-2023", "dataset_10-2022","SpotifyAudioFeatures_04-2019","SpotifyAudioFeatures_11-2018"]
```

```python
# Function to extract year and month from the filename
def extract_year_month(filename):
    parts = filename.split('_')
    month = parts[1][:2]
    year = parts[1][3:7]
    return month, year

# Load CSV files and add year_data and month_data columns
dataframes = []
for name in names_data:
    df = pd.read_csv(name + '.csv')
    month, year = extract_year_month(name)
    df.rename(columns={'popularity': f'popularity_{year}'}, inplace=True)
    dataframes.append(df)

# Combine all dataframes into one
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df
```

```python
missing_track_id_count = combined_df['track_id'].isnull().sum()
print(f"Missing track_id count: {missing_track_id_count}")

combined_df.describe()

```
