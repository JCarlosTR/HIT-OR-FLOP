```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.dtypes
```

```python
df["imputed_date"] = (df["release_date"].str.match(r"^\d{4}$")) | (
    df["release_date"].str.match(r"^\d{4}-\d{2}$")
)
```

```python
df["id"] = df["id"].astype("string")
df["key"] = df["key"].astype("int8")
df["mode"] = df["mode"].astype("UInt8")
df["explicit"] = df["explicit"].astype("UInt8")
df["name"] = df["name"].astype("string")
df["popularity"] = df["popularity"].astype("UInt8")
df["release_date"] = df["release_date"].astype("string")
df["year"] = df["year"].astype("UInt16")
df["duration_ms"] = df["duration_ms"].astype("UInt64")
df["artists"] = df["artists"].astype("string")

df["artists"] = (df["artists"].str.replace(r"[\[\]']", "", regex=True).str.replace(",", ", "))

df.loc[df["release_date"].str.match(r"^\d{4}$"), "release_date"] += "-01-01"
df.loc[df["release_date"].str.match(r"^\d{4}-\d{2}$"), "release_date"] += "-01"

df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce", format="%Y-%m-%d")

df["month"] = df["release_date"].dt.month

df["featuring"] = df["artists"].str.contains(",", regex=False)
```

```python
print(df.dtypes)
```

```python
len(df[df.duplicated(keep=False) == True])
```

```python
df = df[df.duplicated(keep=False) == False]
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv("data_formatted.csv", index=False)
```
