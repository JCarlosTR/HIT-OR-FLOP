```python
import pandas as pd
```

```python
df_w_songs_genres = pd.read_csv("canciones_generos_y_subgeneros_separados_2024.csv")

df_w_songs_genres.rename(columns={'track_id': 'id'}, inplace=True)


df = df_w_songs_genres.copy()

df
```

```python
df["imputed_date"] = (df["release_date"].str.match(r"^\d{4}$")) | (
    df["release_date"].str.match(r"^\d{4}-\d{2}$")
)

df.loc[df["release_date"].str.match(r"^\d{4}$"), "release_date"] += "-01-01"
df.loc[df["release_date"].str.match(r"^\d{4}-\d{2}$"), "release_date"] += "-01"
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce", format="%Y-%m-%d")

df["month"] = df["release_date"].dt.month.astype('UInt16')
df["year"] = df["release_date"].dt.year.astype("UInt16")



df["featuring"] = df["artist_name"].str.contains(",", regex=False)
```

```python
df
```

```python
df_category = df[['artist_name','track_name']].copy()

df_w_songs_genres = df.copy()

df_w_songs_genres
```

```python
def categorize_popularity(value):
    if value <= 25:
        return "FLOP"
    elif 25 < value <= 50:
        return "MID"
    elif 50 < value <= 80:
        return "MID-HIT"
    
    return "HIT"

def categorize_numeric_features(value):
    if value <= 0.5:
        return 'Low'
    else:
        return 'High'

def categorize_bool_features(value):
    if value == 1 or value == True:
        return 'Yes'
    else:
        return 'No'

def categorize_mode(value):
    if value == 1:
        return 'Major'
    else:
        return 'Minor'


def categorize_key(value):
    key_dict = {
        -1: "Unknown",
        0: "C",
        1: "C♯/D♭",
        2: "D",
        3: "D♯/E♭",
        4: "E",
        5: "F",
        6: "F♯/G♭",
        7: "G",
        8: "G♯/A♭",
        9: "A",
        10: "A♯/B♭",
        11: "B",
    }
    return key_dict[value]

def categorize_duration(value):
    duration_dict = {
        (0, 60000): "Very short",
        (60001, 120000): "Short",
        (120001, 300000): "Mid",
        (300001, 480000): "Large",
        (480000, float("inf")): "Very large"
    }
    
    for (low, high), category in duration_dict.items():
        if low <= value < high:
            return category

def categorize_tempo(value):
    tempo_dict = {
        (0, 20): "Sin Tempo",
        (20, 40): "Grave",
        (40, 60): "Lento",
        (60, 66): "Largo",
        (66, 70): "Larghetto",
        (70, 76): "Adagio",
        (76, 80): "Adagietto",
        (80, 108): "Andante",
        (108, 112): "Moderato",
        (112, 120): "Allegretto",
        (120, 168): "Allegro",
        (168, 176): "Vivace",
        (176, 200): "Presto",
        (200, float("inf")): "Prestissimo"
    }

    for (low, high), category in tempo_dict.items():
        if low <= value < high:
            return category
    
    return 

def categorize_year(value):
    decade = (value // 10) * 10
    if value < 2000:
        return f"{decade % 100}s"
    else:
        return f"{decade}s"
```

```python
numeric_columns = [col for col in df_w_songs_genres.columns if col not in ["artist_name","duration_ms","id","key","mode","track_name","release_date","year","imputed_date","album"] and "genre" not in col and "popularity" not in col]

df_w_songs_genres["release_date"] = pd.to_datetime(df_w_songs_genres["release_date"])
```

```python
for col in numeric_columns:
    df_w_songs_genres[col + "_cat"] = df_w_songs_genres[col].apply(categorize_numeric_features).astype('category')

df_w_songs_genres['popularity_2024_cat'] = df_w_songs_genres['popularity_2024'].apply(categorize_popularity).astype('category')
df_w_songs_genres['popularity_2025_cat'] = df_w_songs_genres['new_popularity_2025'].apply(categorize_popularity).astype('category')

df_w_songs_genres['duration_cat'] = df_w_songs_genres['duration_ms'].apply(categorize_duration).astype('category')

df_w_songs_genres["tempo_cat"] = (
    df_w_songs_genres["tempo"].apply(categorize_tempo).astype("category")
)

df_w_songs_genres["key_cat"] = df_w_songs_genres["key"].apply(categorize_key).astype("category")

df_w_songs_genres['explicit_cat'] = df_w_songs_genres['explicit'].apply(categorize_bool_features).astype('category')
df_w_songs_genres['featuring_cat'] = df_w_songs_genres['featuring'].apply(categorize_bool_features).astype('category')
df_w_songs_genres['genres_from_artists_cat'] = df_w_songs_genres['genres_from_artists'].apply(categorize_bool_features).astype('category')


df_w_songs_genres["imputed_date_cat"] = (
    df_w_songs_genres["imputed_date"].apply(categorize_bool_features).astype("category")
)

df_w_songs_genres["mode_cat"] = (
    df_w_songs_genres["mode"].apply(categorize_mode).astype("category")
)

df_w_songs_genres["month_cat"] = (
    df_w_songs_genres["release_date"].dt.month_name().astype("category")
)

df_w_songs_genres['year_cat'] = df_w_songs_genres['year'].apply(categorize_year).astype('category')
```
