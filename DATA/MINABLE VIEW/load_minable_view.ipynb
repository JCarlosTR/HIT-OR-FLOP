```python
import numpy as np
import pandas as pd


def categorize_popularity(value):
    if value <= 25:
        return "FLOP"
    elif 25 < value <= 50:
        return "MID"
    elif 50 < value <= 80:
        return "MID-HIT"
    
    return "HIT"

df_21 = pd.read_csv('data_genres_2025_21.csv')

df_21 = df_21.drop_duplicates(subset=['artist_name', "track_name"], keep='first')
df_21 = df_21.drop(columns=['artists', "name"])

df_21['popularity_2025_cat'] = df_21['new_popularity_2025'].apply(categorize_popularity).astype('category') # añadimos la categórica de 2025
df_21.rename(columns={"popularity_cat": "popularity_2021_cat"}, inplace=True)
df_21["album"] = "unknown"
df_21["popularity_2024"] = "unknown"
df_21["popularity_2024_cat"] = "unknown"


df_21.rename(columns={"parsed_genres_artists": "genres_artists"}, inplace=True)
df_21
```

```python
# fix featuring cat
df_21["featuring_cat"] = df_21["artist_name"].str.contains(",", regex=False)
df_21["featuring_cat"] = df_21["featuring_cat"].replace({True: "Yes", False: "No"})

featuring_cat_counts = df_21["featuring_cat"].value_counts(dropna=False)
print(featuring_cat_counts)
```

```python
df_24 = pd.read_csv("data_genres_2025_24.csv")
df_24 = df_24.drop_duplicates(subset=['artist_name', "track_name"], keep='first')
df_24.drop(columns=['imputed_date', 'genres_from_artists', 'month', 'featuring', "genres_artists"], inplace=True)
df_24["popularity_2021"] = "unknown"
df_24["popularity_2021_cat"] = "unknown"
df_24['year_cat'] = df_24['year_cat'].str.replace('.0s', 's', regex=False)
df_24.rename(columns={"parsed_genres_artists": "genres_artists"}, inplace=True)
df_24
```

```python
# Filtrar filas donde genres_artists no sea "[]" o "[unchecked]"
filtered_df = df_24

# Contar el número de filas por artista
artist_counts = filtered_df['artist_name'].value_counts()

# Buscar artistas con más de una fila
artists_with_multiple_rows = artist_counts[artist_counts > 1].index

# Filtrar el dataframe original para obtener las filas de estos artistas
result_df = df_24[df_24['artist_name'].isin(artists_with_multiple_rows)]

# Filtrar artistas que tienen al menos un genres_artists igual a "[]" o "[unchecked]"
artists_with_unchecked = df_24[df_24['genres_artists'].isin(["[[]]", '[[unchecked]]'])]['artist_name']

# Obtener los artistas que tienen al menos una fila que no es "[]" ni "[unchecked]"
artists_with_valid_genres = df_24[~df_24['genres_artists'].isin(["[[]]", '[[unchecked]]'])]['artist_name']

print("Count of unknown genres rows: ", len(artists_with_unchecked))
print("Count of known genres rows: ", len(artists_with_valid_genres))

# Hallar la intersección de artists_with_unchecked y artists_with_valid_genres
intersection_artists = artists_with_unchecked[artists_with_unchecked.isin(artists_with_valid_genres)]
print("Artists in both sets: ", len(intersection_artists))
```

Filtering high speechiness rows (Podcasts, audiobooks, ...)

```python
from sklearn.preprocessing import StandardScaler

df_copy = df_merged.copy()
# Estandarizar el dataframe
scaler = StandardScaler()
df_copy[df_copy.select_dtypes(include=[np.number]).columns] = scaler.fit_transform(df_copy.select_dtypes(include=[np.number]))

# Convertir speechiness a valor absoluto
df_copy['speechiness'] = df_copy['speechiness'].abs()
speechiness_rows = df_copy[df_copy['speechiness'] > 4]
# Obtener los índices de high_speechiness_rows
indices_speechiness_rows = speechiness_rows.index


# Filtrar el dataframe original para obtener los nombres de las canciones
song_names = df_copy.loc[indices_speechiness_rows, 'track_name']
speechiness_value = df_copy.loc[indices_speechiness_rows, 'speechiness']


filtered_song_names = song_names[song_names.str.contains("pt|jfl|skit|speech|story|announcement|récital|acte|monologue|commentary|chat|lecture|part|catalunya|radio drama|capítulo|chapter|Часть|kapitel|interview", case=False, na=False)]

# Eliminar las filas de merged_df donde track_name está en filtered_song_names
df_merged = df_merged[~df_merged['track_name'].isin(filtered_song_names)]

df_merged
```

```python
# Encontrar las filas con ids duplicados
duplicated_ids = df_merged[df_merged.duplicated(subset='id', keep=False)]

for id_val in duplicated_ids['id'].unique():
    duplicates = duplicated_ids[duplicated_ids['id'] == id_val]
    
    if len(duplicates) > 1:
        # Obtener los índices reales
        idx_first = duplicates.index[0]
        idx_second = duplicates.index[1]
        
        # Sustituir valores "unknown" directamente en df_merged usando .loc
        for col in ['popularity_2021', 'popularity_2021_cat', 'popularity_2024', 'popularity_2024_cat']:
            if duplicated_ids.loc[idx_first, col] == "unknown":
                duplicated_ids.loc[idx_first, col] = duplicated_ids.loc[idx_second, col]
        
        # Eliminar la fila duplicada (segunda)
        duplicated_ids = duplicated_ids.drop(idx_second)

# Mostrar las filas con ids duplicados
duplicated_ids
```
